```python

# 🔹 Associer les croisements aux nodes OSM
croisement_to_node = {
    croisement['nom']: ox.nearest_nodes(G, X=croisement['coord'][1], Y=croisement['coord'][0])
    for croisement in croisements
}
```

```python

# v1 🔹 Fonction pour afficher la séquence des rues entre deux croisements
def afficher_chemin(croisement_depart, croisement_arrivee):
    node_depart = croisement_to_node.get(croisement_depart)
    node_arrivee = croisement_to_node.get(croisement_arrivee)

    if node_depart is None or node_arrivee is None:
        print("❌ Croisement non trouvé.")
        return

    try:
        chemin = nx.shortest_path(G, node_depart, node_arrivee, weight='length')

        print(f"🔵 Chemin de {croisement_depart} à {croisement_arrivee} :")
        previous_edge_name = None

        for i in range(len(chemin)-1):
            u = chemin[i]
            v = chemin[i+1]
            edge_data = G.get_edge_data(u, v)

            # Parfois plusieurs segments possibles entre deux nœuds : on prend le premier
            if edge_data:
                edge_info = list(edge_data.values())[0]
                nom_rue = edge_info.get('name', 'Rue sans nom')

                if nom_rue != previous_edge_name:  # éviter doublons successifs
                    print(f"➡️ {nom_rue}")
                    previous_edge_name = nom_rue
        print("\n")

    except nx.NetworkXNoPath:
        print(f"⚠️ Aucun chemin direct trouvé entre {croisement_depart} et {croisement_arrivee}.\n")

# Exemple d'utilisation :
afficher_chemin("CROIST_OBERKAMPF_VOLTAIRE", "CROIST_OBERKAMPF_SAINT-MAUR")
```

```python
import requests
from geopy.distance import geodesic
coords = [
    {'coord': (48.8686651, 2.3670429), 'nom': 'CROIST_DU_FAUBOURG_DU_TEMPLE_JULES_FERRY'},
    {'coord': (48.8700004, 2.3710483), 'nom': 'CROIST_DU_FAUBOURG_DU_TEMPLE_PARMENTIER'},
    {'coord': (48.8705636, 2.3728843), 'nom': 'CROIST_DU_FAUBOURG_DU_TEMPLE_SAINT-MAUR'},
    {'coord': (48.8637626, 2.3701149), 'nom': 'CROIST_OBERKAMPF_VOLTAIRE'},
    {'coord': (48.8661922, 2.3696437), 'nom': 'CROIST_DE_LA_REPUBLIQUE_RICHARD-LENOIR'},
    {'coord': (48.8656609, 2.3773308), 'nom': 'CROIST_OBERKAMPF_SAINT-MAUR'},
    {'coord': (48.8631027, 2.3667404), 'nom': 'CROIST_DES_FILLES_DU_CALVAIRE_DU_TEMPLE'},
    {'coord': (48.8573512, 2.3808768), 'nom': 'CROIST_RICHARD_LENOIR_VOLTAIRE'},
    {'coord': (48.8646123, 2.3780241), 'nom': 'CROIST_DE_LA_REPUBLIQUE_SAINT-MAUR'},
    {'coord': (48.8639984, 2.3706775), 'nom': 'CROIST_OBERKAMPF_RICHARD-LENOIR'},
    {'coord': (48.8642128, 2.3711835), 'nom': 'CROIST_OBERKAMPF_RICHARD-LENOIR'},
    {'coord': (48.8688224, 2.3676207), 'nom': 'CROIST_DU_FAUBOURG_DU_TEMPLE_JULES_FERRY'},
    {'coord': (48.869535, 2.369313), 'nom': 'A'},
    {'coord': (48.869756, 2.370046), 'nom': 'B'},
    {'coord': (48.870826, 2.370584), 'nom': 'C'},
    {'coord': (48.870052, 2.371149), 'nom': 'D'}
]

def get_directions_real(start_lat, start_lng, end_lat, end_lng, api_key):
    """Appelle l'API Google Directions et retourne les étapes."""
    url = f"https://maps.googleapis.com/maps/api/directions/json?origin={start_lat},{start_lng}&destination={end_lat},{end_lng}&mode=walking&key={api_key}"
    response = requests.get(url)
    data = response.json()

    if data["status"] != "OK":
        print(f"❌ Erreur API: {data['status']}")
        return None

    steps = data["routes"][0]["legs"][0]["steps"]
    etapes = []

    for step in steps:
        rues = []
        instruction = step.get("html_instructions", "")
        distance = step.get("distance", {}).get("value", 0)  # distance en mètres

        # Extraire toutes les rues mentionnées
        if "<b>" in instruction:
            segments = [seg.split("</b>")[0] for seg in instruction.split("<b>")[1:]]
            rues = [s.strip() for s in segments if s.strip()]

        etapes.append({
            'rues': rues,
            'distance': distance
        })

    return etapes


def nettoyer_rues(etapes):
    rues_affichees = set()
    rues_finales = []

    for i, etape in enumerate(etapes):
        rues = etape['rues']
        distance = etape['distance']

        if not rues:
            continue

        # Toujours garder la première rue
        rue_principale = rues[0]
        if rue_principale not in rues_affichees:
            rues_finales.append(rue_principale)
            rues_affichees.add(rue_principale)

        # Tester une éventuelle seconde rue
        if len(rues) > 1:
            rue_secondaire = rues[1]

            deja_vue = rue_secondaire in rues_affichees
            apparait_ailleurs = any(
                rue_secondaire in e['rues'] for j, e in enumerate(etapes) if j != i
            )

            if distance >= 5 or apparait_ailleurs:
                if not deja_vue:
                    rues_finales.append(rue_secondaire)
                    rues_affichees.add(rue_secondaire)
            else:
                print(f"❌ Rue suspecte ignorée : {rue_secondaire} (distance {distance} m)")

    return rues_finales

```

```python
def afficher_chemin(croisement_depart, croisement_arrivee, api_key):
    start = next((item['coord'] for item in coords if item['nom'] == croisement_depart), None)
    end = next((item['coord'] for item in coords if item['nom'] == croisement_arrivee), None)

    if not start or not end:
        return "❌ Croisement non trouvé."

    etapes = get_directions_real(start[0], start[1], end[0], end[1], api_key)
    if not etapes:
        return "❌ Aucune étape trouvée."

    texte = f"🔍 Résultat brut de get_directions pour {croisement_depart} ➡️ {croisement_arrivee} :\n"
    for e in etapes:
        texte += f"➡️ {e}\n"

    rues = nettoyer_rues(etapes)

    texte += f"\n🔵 Chemin final de {croisement_depart} à {croisement_arrivee} :\n"
    for rue in rues:
        texte += f"➡️ {rue}\n"

    return texte
```
